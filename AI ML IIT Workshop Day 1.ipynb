```python
g={"Name":['a','b','c','d','e'],'CGPA':[7,9,6,8,5]}
```

```python
pt.pie(g)
```

```python
#giving a title
pt.title("CGPA of the students")
pt.pie(g['CGPA'])
```

```python
#also gives names of the students i.e. a,b,c,d,e
pt.title("CGPA of the students")
pt.pie(g['CGPA'],labels=g['Name'])
pt.show
```

```python
#Only take out a seperately
pt.title("CGPA of the students")
pt.pie(g['CGPA'],labels=g['Name'],explode=[1,0,0,0,0])
pt.show
```

```python
# get to customize the colors of the pie chart
pt.title("CGPA of the students")
pt.pie(g['CGPA'],labels=g['Name'],explode=[0.2,0,0,0,0],colors=['blue','red','green','pink','yellow'])
pt.show
```

```python
# auto % calculates the value of each pie : %f => float ; %.2f => 2 decimal places
pt.title("CGPA of the students")
pt.pie(g['CGPA'],labels=g['Name'],explode=[0.2,0,0,0,0],colors=['blue','red','green','pink','yellow'],autopct="%f")
pt.show
```

```python
pt.title("CGPA of the students")
pt.pie(g['CGPA'],labels=g['Name'],explode=[0.2,0,0,0,0],colors=['blue','red','green','pink','yellow'],autopct="%.2f")
pt.show
```
